```python
class Solution:
    def calculate(self, s: str) -> int:
        if not(s):
            return 0
        elif s[0] == '-':
            s = '0' + s

        lst_s = []
        curr_num = None
        for w in s:
            if w == ' ':
                continue
            elif w.isdigit():
                if not(curr_num):
                    curr_num = w
                else:
                    curr_num += w
            elif w in ['-', '+',')']:
                if not(curr_num) and (not(lst_s) or lst_s[-1]) != ')':
                    curr_num = w
                elif not(curr_num):
                    lst_s += [w]
                else:
                    lst_s += [curr_num]
                    curr_num = None
                    lst_s += [w]
            elif w == '(' and lst_s and curr_num == '-' and lst_s[-1] == '(':
                lst_s += ['0']
                lst_s += ['-']
                lst_s += [w]
                curr_num = None
            else:
                lst_s += [w]
        if curr_num:
            lst_s += [curr_num]   

        print(lst_s)
        res = 0    
        stack_s = []
        
        
        for w in lst_s:
            if w == '(':
                stack_s += [w]
            elif w.isdigit():
                if not(stack_s):
                    stack_s += [int(w)]
                elif stack_s[-1] in ['+', '-']:
                    tempsum = int(w)
                    while stack_s and stack_s[-1] in ['+', '-']:
                        x = stack_s.pop()
                        y = stack_s.pop()
                        if x == '+':
                            tempsum += int(y)
                        else:
                            tempsum = int(y) - tempsum
                    stack_s += [tempsum]
                else:
                    stack_s += [int(w)]
            elif w == ')':
                x = stack_s.pop()
                y = stack_s.pop()
                tempsum = int(x)
                while stack_s and stack_s[-1] in ['+', '-']:
                    x = stack_s.pop()
                    y = stack_s.pop()
                    if x == '+':
                        tempsum += int(y)
                    else:
                        tempsum = int(y) - tempsum
                stack_s += [tempsum]
            else:
                stack_s += [w]

        return stack_s.pop()
```
