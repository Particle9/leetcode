```python
class Solution:
    def getStringDiff(self, s1,s2):
        sumDiff = 0
        for c1,c2 in zip(s1,s2):
            if c1 != c2:
                sumDiff += 1
        return sumDiff
    
    def getMutation(self, curr, end, graph, visited):
        if curr == end:
            return 0
        
        visited.add(curr)
        minMutation = float('inf')
        for x in graph[curr]:
            if x not in visited:
                minMutation = min(minMutation, self.getMutation(x, end, graph, visited) + 1) 
        
        return minMutation


    def minMutation(self, startGene: str, endGene: str, bank: List[str]) -> int:
        graph = {startGene:[]}
        for b in bank:
            lstKey = [k for k in graph]
            for k in lstKey:
                if self.getStringDiff(b,k) == 1:
                    graph[k] += [b]
                    if b not in graph:
                        graph[b] = [k]
                    else:
                        graph[b] += [k]
        
        for k,lv in graph.items():
            graph[k] = sorted(lv, key= lambda x: self.getStringDiff(x,endGene))
        visited = set()
        countMutation = 0
        queue = [(startGene, 0)]
        
        while queue:
            x, n = queue.pop(0)
            if x == endGene:
                return n
            elif x not in visited:
                visited.add(x)
                queue += [(y, n+1) for y in graph[x]]
        
        return -1


        # countMutation = self.getMutation(startGene, endGene, graph, visited)
        # if countMutation == float('inf'):
        #     return -1
        # return countMutation
                    

        
```
