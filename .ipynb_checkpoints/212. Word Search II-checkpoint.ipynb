```python

class Solution:
    def isPrefix(self, sW, lW):
        if len(sW) > len(lW):
            return False
        return sW == lW[:len(sW)]

    def findPrefix(self, word, wordDict):
        for kw in reversed(wordDict.keys()):
            if self.isPrefix(kw, word):
                return kw
        return ""

    def explodeWords(self, words):
        setWords = set()
        for w in words:
            for i in range(0,len(w)):
                wx = w[:len(w)-i]
                if wx in setWords:
                    break
                setWords.add(wx)
        return sorted(setWords,key=len)
    
    def wordSearch(self, i, j, board, visited, trie, currWord, wordBanks):
        if not(trie[currWord]) :
            return [currWord]
        possWords = trie[currWord]

        if currWord in wordBanks:
            wordsFound = [currWord]
        else:
            wordsFound = []

        visitedBoard = visited | {(i,j)}
        for t in [(i+1,j), (i-1,j), (i,j+1), (i,j-1)]:
            x,y = t
            if t not in visited and 0 <= x < len(board) and 0 <= y < len(board[0]):
                newWord = currWord + board[x][y]
                if newWord in trie:
                    wordsFound += self.wordSearch(x,y,board, visitedBoard, trie, newWord, wordBanks)
        return wordsFound

    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        trie = {"":[]}
        allWords = self.explodeWords(words)

        for w in allWords:
            kw = self.findPrefix(w,trie)
            trie[kw] += [w]
            trie[w] = []
        
        wordBanks = set(words)
        possLst = []
        for i in range(0, len(board)):
            for j in range(0,len(board[0])):
                if board[i][j] in trie:
                    visited = set(board[i][j])
                    possLst += self.wordSearch(i, j, board, visited, trie, board[i][j], wordBanks)
        
        return list(set(possLst))


        
```
