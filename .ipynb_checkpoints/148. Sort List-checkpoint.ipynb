```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def quickSort(self, lst):
        if not(lst):
            return []
        elif len(lst) == 1:
            return lst
        
        i_mid = len(lst) //2
        lst_left = self.quickSort(lst[:i_mid])
        lst_right = self.quickSort(lst[i_mid:])
        i,j = 0,0

        tempLst = []
        while i < len(lst_left) and j < len(lst_right):
            if lst_left[i].val > lst_right[j].val:
                tempLst += [lst_right[j]]
                j+=1
            else:
                tempLst += [lst_left[i]]
                i+= 1
        
        if j < len(lst_right):
            tempLst += lst_right[j:]
        else:
            tempLst += lst_left[i:]
        
        return tempLst



    def sortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not(head):
            return None
        elif head.next == None:
            return ListNode(head.val)
        
        nextElmt = head.next
        # tempLst = [min(head, nextElmt, key=lambda x: x.val), max(head, nextElmt, key=lambda x: x.val)]
        tempLst = [head]
        while nextElmt != None:    
            tempLst = tempLst + [nextElmt]
            nextElmt = nextElmt.next
        tempLst = self.quickSort(tempLst)
        
        tempLst += [None]
        for i in range(0,len(tempLst)-1):
            tempLst[i].next = tempLst[i+1]
        return tempLst[0]

                
        
        
```
