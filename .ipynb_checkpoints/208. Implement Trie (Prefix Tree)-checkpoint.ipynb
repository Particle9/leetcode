```python
def isPrefix(sW,lW):
    if len(sW) > len(lW):
        return False
    for c1,c2 in zip(sW,lW):
        if c1 != c2:
            return False
    return True


class Trie:

    def __init__(self):
        self.tree = {"":[]}
        self.sortedNodes = []
        

    def insert(self, word: str) -> None:
        if word not in self.tree:
            isPlacedInNode = False
            isPlacedInList = False
            for i in range(0, len(self.sortedNodes)):
                if isPrefix(self.sortedNodes[i], word) and not(isPlacedInNode):
                    self.tree[self.sortedNodes[i]] += word
                    self.tree[word] = []
                    isPlacedInNode = True
                if isPrefix(word, self.sortedNodes[i]) and not(isPlacedInNode):
                    self.tree[word] = [self.sortedNodes[i]]
                    isPlacedInNode = True
                if len(word) > len(self.sortedNodes[i]) and not(isPlacedInList):
                    self.sortedNodes = self.sortedNodes[:i] + [word] + self.sortedNodes[i:]
                    isPlacedInList = True
            if word not in self.tree:
                self.tree[word] = []
                self.tree[""] += [word]
            if not(isPlacedInList):
                self.sortedNodes += [word]

    def search(self, word: str) -> bool:
        return word in self.tree  

    def startsWith(self, prefix: str) -> bool:
        if prefix in self.tree:
            return True

        for x in self.sortedNodes:
            if isPrefix(prefix,x):
                return True

        return False
        


# Your Trie object will be instantiated and called as such:
# obj = Trie()
# obj.insert(word)
# param_2 = obj.search(word)
# param_3 = obj.startsWith(prefix)
```
