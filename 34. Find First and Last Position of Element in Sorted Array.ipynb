```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        if not(nums):
            return [-1,-1]
        elif len(nums) == 1:
            return [0,0] if nums[0] == target else [-1,-1] 
                
        lower, upper = 0, len(nums) -1
        targetIdx = -1
        while lower < upper:
            mid = (upper + lower)//2
            if nums[mid] == target:
                targetIdx = mid
                break
            elif nums[upper] == target:
                targetIdx = upper
                break
            elif nums[lower] == target:
                targetIdx = lower
                break
            
            if nums[mid] > target:
                upper = mid-1
            else:
                lower = mid+1

        if targetIdx == -1:
            return [-1,-1]
        
        lowIdx = targetIdx
        lower = 0
        upper = targetIdx
        while lower < upper:
            mid = (upper + lower) // 2
            if nums[mid] == target and nums[mid-1] < target:
                lowIdx = mid
                break
            elif nums[lower] == target and (lower == 0 or nums[lower-1] < target):
                lowIdx = lower
                break
            
            if nums[mid] == target:
                upper = mid - 1
            else:
                lower = mid + 1

        upIdx = targetIdx
        lower = targetIdx
        upper = len(nums) - 1
        while lower < upper:
            mid = (upper + lower) // 2
            if nums[mid] == target and nums[mid+1] > target:
                upIdx = mid
                break
            elif nums[upper] == target and (upper == len(nums)-1 or nums[upper+1] > target):
                upIdx = upper
                break
            if nums[mid] == target:
                lower = mid + 1
            else:
                upper = mid - 1
        return [lowIdx, upIdx]

        
        
```
